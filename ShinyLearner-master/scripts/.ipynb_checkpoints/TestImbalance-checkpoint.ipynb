# Get subset as numpy arrays

#### X and y get change in a function, then we write them

```python
import pandas as pd
import numpy as np
import sys

DataFilePath = "../Validation/StrongSignal_Continuous.tsv.gz"

def Sampling(DataFilePath, SamplingMethod, OutputFilePath):
    

    df = pd.read_csv(DataFilePath, sep = '\t', index_col = 0)


    X = df.iloc[:,:-1].values.astype(np.float32)
    X_columns = df.iloc[:,:-1].columns

    y = df.iloc[:,-1:].values.astype(np.int32).ravel()
    y_columns = df.iloc[:,-1:].columns
    
    if SamplingMethod == "NearMiss":
        from imblearn.under_sampling import NearMiss
        sampler = NearMiss(version=2)
    elif SamplingMethod == "ClusterCentroids":
        from imblearn.under_sampling import ClusterCentroids
        sampler = ClusterCentroids()
    elif SamplingMethod == "RandomOverSampler":
        from imblearn.over_sampling import RandomOverSampler
        sampler = RandomOverSampler()
    elif SamplingMethod == "ADASYN":
        from imblearn.over_sampling import ADASYN
        sampler = ADASYN()

    X_resampled, y_resampled = sampler.fit_sample(X, y)
    
    indices = []
    count = 1
    for line in X_resampled:
        indices.append("Sample__ID__" + str(count))
        count +=1
    
    X_Final = pd.DataFrame(data=X_resampled,columns=X_columns)
    y_Final = pd.DataFrame(data=y_resampled,columns=y_columns)

    
    X_y_Final = pd.concat([X_Final,y_Final], axis=1)
    
    X_y_Final.index = indices
    
    X_y_Final.to_csv(OutputFilePath, sep='\t')
    print X_y_Final

Sampling(DataFilePath, "NearMiss", 'ImbalanceResults/output_NearMiss.tsv')
Sampling(DataFilePath, "ClusterCentroids",'ImbalanceResults/output_ClusterCentroids.tsv' )
Sampling(DataFilePath, "RandomOverSampler", 'ImbalanceResults/output_RandomOverSampler.tsv')
Sampling(DataFilePath, "ADASYN", 'ImbalanceResults/output_ADASYN.tsv')
```
